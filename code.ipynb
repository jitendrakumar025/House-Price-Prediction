```python
import pandas as pd
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("hackothan.csv")
```

```python
df
```

```python
df.drop("index", axis =1, inplace=True)
df
```

```python
target = df.retail_price
target
```

```python
df.drop("retail_price", axis =1, inplace=True)
df
```

```python
adress = df["address"].unique()
adress
```

```python
adress.shape
```

```python
city = df["city"].unique()
city.shape
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
```

```python
df["city"]=label_encoder.fit_transform(df["city"])
df
```

```python
df["address"]=label_encoder.fit_transform(df["address"])
df
```

```python
df.columns[df.isna().any()]
df.rooms = df.rooms.fillna(df.rooms.mean())
df.sqft = df.sqft.fillna(df.sqft.mean())
```

```python
inputs= df
```

```python
from sklearn.preprocessing import MinMaxScaler
inputs["address"]=MinMaxScaler().fit_transform(np.array(inputs["address"]).reshape(-1,1))
```

```python
inputs["sqft"]=MinMaxScaler().fit_transform(np.array(inputs["sqft"]).reshape(-1,1))
inputs["image_id"]=MinMaxScaler().fit_transform(np.array(inputs["image_id"]).reshape(-1,1))
inputs["city"]=MinMaxScaler().fit_transform(np.array(inputs["city"]).reshape(-1,1))
inputs
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(inputs, target, test_size=0.2)
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 200, random_state = 0)
regressor.fit(X_train, y_train)
```

```python
regressor.score(X_train, y_train)
```
